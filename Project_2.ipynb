```python
# Step 2: Define the path to the zip file and the extraction directory
zip_file_path = '/content/drive/MyDrive/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone.zip'
extract_dir = '/content/CT-KIDNEY-DATASET'

# Step 3: Create the extraction directory if it doesn't exist
if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

# Step 4: Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f'Dataset extracted to {extract_dir}')
```

```python
import os
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Constants for image and model configurations
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
INPUT_SHAPE = (224, 224, 3)
NUM_CLASSES = 4  # Classes: Normal, Cyst, Tumor, Stone

# Function to create dataframe from directory
def create_dataframe(data_dir):
    """
    Creates a dataframe with image paths and corresponding labels.
    """
    images = []
    labels = []
    for label in os.listdir(data_dir):
        label_dir = os.path.join(data_dir, label)
        if os.path.isdir(label_dir):
            for img_name in os.listdir(label_dir):
                img_path = os.path.join(label_dir, img_name)
                if os.path.isfile(img_path) and img_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif', '.tiff')):
                    images.append(img_path)
                    labels.append(label)
    df = pd.DataFrame({'image_path': images, 'labels': labels})
    return df

# Function to create ImageDataGenerator
def create_image_data_generator(df, batch_size=32, mode='training'):
    """
    Creates an ImageDataGenerator and flows data from a dataframe.
    """
    if mode == 'training':
        datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=30,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='nearest'
        )
        shuffle = True
    elif mode in ['validation', 'testing']:
        datagen = ImageDataGenerator(rescale=1./255)
        shuffle = False  # Ensure no shuffling for validation and testing
    else:
        raise ValueError("mode should be 'training', 'validation', or 'testing'")

    if mode in ['training', 'validation']:
        data_generator = datagen.flow_from_dataframe(
            dataframe=df,
            x_col='image_path',
            y_col='labels',
            target_size=IMAGE_SIZE,
            batch_size=batch_size,
            class_mode='categorical',
            shuffle=shuffle  # Shuffle only if mode is training
        )
    elif mode == 'testing':
        data_generator = datagen.flow_from_dataframe(
            dataframe=df,
            x_col='image_path',
            y_col=None,
            target_size=IMAGE_SIZE,
            batch_size=batch_size,
            class_mode=None,
            shuffle=False
        )
    return data_generator

# Paths to directories
data_path = "/content/CT-KIDNEY-DATASET/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone"

# Create dataframe
df = create_dataframe(data_path)
print(f"Total images found: {len(df)}")

# Check if dataset is not empty
if df.empty:
    raise ValueError("The dataframe is empty. Please check your data directory path and ensure it contains image files.")

# Split dataframe into training and validation (70% train, 30% validation)
train_df, valid_df = train_test_split(df, test_size=0.3, stratify=df['labels'], random_state=42)
print(f"Training samples: {len(train_df)}")
print(f"Validation samples: {len(valid_df)}")

# Create data generators
train_generator = create_image_data_generator(train_df, batch_size=BATCH_SIZE, mode='training')
valid_generator = create_image_data_generator(valid_df, batch_size=BATCH_SIZE, mode='validation')

# Check class indices
class_indices = train_generator.class_indices
print("Class indices:", class_indices)

# Load the MobileNetV2 model, excluding the top layers
base_model = tf.keras.applications.MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=INPUT_SHAPE
)

# Unfreeze the entire base model for training
base_model.trainable = True

# Build the model
model = Sequential([
    Input(shape=INPUT_SHAPE),
    base_model,
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(1024, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(NUM_CLASSES, activation='softmax')
])

# Compile the model
optimizer = Adam(learning_rate=1e-5)  # Lower learning rate for fine-tuning
model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()

# Callbacks: early stopping and reduce learning rate on plateau
reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    patience=2,
    verbose=1,
    factor=0.5,
    min_lr=1e-6
)
early_stop = EarlyStopping(
    monitor='val_accuracy',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

# Train the model (including fine-tuning)
history = model.fit(
    train_generator,
    validation_data=valid_generator,
    epochs=15,
    verbose=1,
    callbacks=[reduce_lr, early_stop]
)

# -------------------- Evaluation --------------------

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from tensorflow.keras.preprocessing.image import load_img, img_to_array

# Use validation set as test set
test_df = valid_df.copy()
test_generator = create_image_data_generator(test_df, batch_size=BATCH_SIZE, mode='validation')

# Predict on test data
predictions = model.predict(test_generator, verbose=1)
predicted_classes = np.argmax(predictions, axis=1)

# True labels
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Compute Confusion Matrix
cm = confusion_matrix(true_classes, predicted_classes)
print("Confusion Matrix:")
print(cm)

# Classification Report
report = classification_report(true_classes, predicted_classes, target_names=class_labels)
print("Classification Report:")
print(report)

# Overall Accuracy
accuracy = accuracy_score(true_classes, predicted_classes)
print(f"Overall Accuracy: {accuracy:.4f}")

# Plot Confusion Matrix
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix')
plt.show()

# Normalized Confusion Matrix (Optional)
cm_normalized = confusion_matrix(true_classes, predicted_classes, normalize='true')

plt.figure(figsize=(8,6))
sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
            xticklabels=class_labels, yticklabels=class_labels)
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Normalized Confusion Matrix')
plt.show()

# Display 10 Random Sample Predictions
num_samples = 10
random.seed(42)  # For reproducibility
sample_indices = random.sample(range(len(true_classes)), num_samples)

for idx in sample_indices:
    img_path = test_df.iloc[idx]['image_path']
    true_label = test_df.iloc[idx]['labels']
    predicted_label = class_labels[predicted_classes[idx]]

    # Load and display the image
    img = load_img(img_path, target_size=IMAGE_SIZE)
    plt.figure(figsize=(4,4))
    plt.imshow(img)
    plt.title(f"True: {true_label} | Predicted: {predicted_label}")
    plt.axis('off')
    plt.show()
```
